```python
import glob
import cv2
import numpy as np
import sys
import time
```

```python
def main():

	images = file_reader()
	porosities = []

	start = time.time()

	for i in range(0, len(images), 3):
		print(i)
		shape = shape_outliner(images[i])
		total_img_pixels = np.count_nonzero(images[i])
		if total_img_pixels != 0:
			total_shape_pixels = np.count_nonzero(shape)
			porosities.append(por_calc(total_img_pixels, total_shape_pixels))

	porosities = np.array(porosities)

	end = time.time()
	print("TIME: " + str(end - start))

	print(np.average(porosities))
```

```python
# Reads in image files specified by the users
def file_reader():
	while True:
		#file_location = "H:\ElkPointGroup\CT scan\16133\B1\VOI(1)"
		#input("Please specify the file path to where your images are stored: ")
		#file_type = "bmp" #input("Please specify the file type. I.e bmp: ")
		files = (glob.glob("H:/ElkPointGroup/CT scan/133/C3/VOI" + "/*.bmp"))
		images = []
		for img_file in files:
			images.append(cv2.imread(img_file, 0))
		if len(images) == 0:
			sys.exit("The folder location or file type you selected were invalid, please try again")
			continue
		break
	return images
```

```python
# Absolutely brute force implementation that transposes the outline or shape of our data set onto a blank array that
# can be referenced as an alternative to assuming all objects are circular in form. Essentially carves the general shape
# Into a new 3D array! 
def shape_outliner(image):

	width = image.shape[0]
	height = image.shape[1]
	shape = np.ones((width, height), dtype=int)
	cur_i = image

	# Move left to right
	for j in range(0, height):
		width_index = width - 1
		height_index = j
		cur_p = cur_i[width_index, height_index]
		while cur_p == 0 and width_index > 0:
			width_index -= 1
			cur_p = cur_i[width_index, height_index]
			shape[width_index, height_index] = 0

	# Move right to left
	for j in range(0, height):
		width_index = 0
		height_index = j
		cur_p = cur_i[width_index, height_index]
		while cur_p == 0 and width_index < width - 1:
			width_index += 1
			cur_p = cur_i[width_index, height_index]
			shape[width_index, height_index] = 0

	shape = cv2.medianBlur(shape.astype(np.float32), 5)

	return shape
```

```python
def por_calc(bright_pixels, total_pixels):
	return (1 - (bright_pixels/float(total_pixels))) * 100
```
